## ⭐ Task 3 — Quartiles & Five-Number Summary

**Dataset:** Marks of 12 students:
```python
marks = [45, 52, 58, 61, 67, 72, 76, 80, 85, 91, 95, 99]
```

**Your job:**
1. Find **Q1**, **Q2 (median)**, **Q3**
2. Calculate the **IQR** (Q3 − Q1)
3. Print the full **five-number summary** (Min, Q1, Q2, Q3, Max)

**Expected output:**
```
=== Five-Number Summary ===
Min    : 45
Q1     : 60.25
Median : 74.0
Q3     : 85.75
Max    : 99
IQR    : 25.5
```

> 💡 **Hints:**
> - `np.percentile(data, 25)` → Q1
> - `np.percentile(data, 50)` → Q2
> - `np.percentile(data, 75)` → Q3
> - IQR = Q3 − Q1

````python
import numpy as np

# Dataset: Marks of 12 students
marks = [45, 52, 58, 61, 67, 72, 76, 80, 85, 91, 95, 99]

# Calculate Q1 (25th percentile), Q2 (50th percentile/median), and Q3 (75th percentile)
q1 = np.percentile(marks, 25)
q2 = np.percentile(marks, 50)  # This is the median
q3 = np.percentile(marks, 75)

# Calculate IQR (Interquartile Range)
iqr = q3 - q1

# Find minimum and maximum
min_val = np.min(marks)
max_val = np.max(marks)

# Print the five-number summary and IQR
print("=== Five-Number Summary ===")
print(f"Min    : {min_val}")
print(f"Q1     : {q1:.2f}")
print(f"Median : {q2:.1f}")
print(f"Q3     : {q3:.2f}")
print(f"Max    : {max_val}")
print(f"\nIQR    : {iqr:.2f}")

# Additional analysis for better understanding
print("\n" + "📊 Statistical Analysis:")
print("-" * 40)
print(f"Data sorted: {marks}")
print(f"Number of students: {len(marks)}")
print(f"\nRange: {max_val - min_val}")
print(f"Lower fence (for outliers): {q1 - 1.5 * iqr:.2f}")
print(f"Upper fence (for outliers): {q3 + 1.5 * iqr:.2f}")

# Check for potential outliers
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr
outliers = [x for x in marks if x < lower_fence or x > upper_fence]
if outliers:
    print(f"Potential outliers: {outliers}")
else:
    print("No outliers detected in this dataset")

# Calculate additional statistics
mean_val = np.mean(marks)
std_val = np.std(marks, ddof=1)  # Sample standard deviation

print(f"\nMean   : {mean_val:.2f}")
print(f"Std Dev: {std_val:.2f}")
print(f"Skewness: {'Positive' if mean_val > q2 else 'Negative' if mean_val < q2 else 'Symmetric'}")
````

## ⭐ Task 4 — Detect Outliers Using IQR

**Dataset:** Annual salaries at a small company (one person is the CEO!):
```python
salaries = [30000, 32000, 35000, 33000, 200000, 31000, 34000, 36000, 29000, 32500]
```

**Your job:**
1. Compute Q1, Q3, and IQR
2. Compute the **outlier fences**:
   - Lower fence = Q1 − 1.5 × IQR
   - Upper fence = Q3 + 1.5 × IQR
3. Loop through salaries and **print any outliers**

**Expected output:**
```
Lower fence : 22,875.00
Upper fence : 43,625.00
🚨 Outlier detected: 200,000
```

> 💡 **Hints:**
> - Use `for s in salaries:` loop
> - Check: `if s < lower or s > upper:`
> - 200,000 (the CEO salary) should be the outlier
> - Use `f'{s:,}'` to format numbers with commas

````python
# Task 4 — Your code here
salaries = [30000, 32000, 35000, 33000, 200000,
            31000, 34000, 36000, 29000, 32500]

# Sort the data for quartile calculation
sorted_salaries = sorted(salaries)
n = len(sorted_salaries)

# Calculate Q1, Q3, and IQR
# Q1 is the median of the first half
# Q3 is the median of the second half
q1_index = n // 4
q3_index = (3 * n) // 4
Q1 = sorted_salaries[q1_index]
Q3 = sorted_salaries[q3_index]
IQR = Q3 - Q1

# Outlier fences
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Print fences with comma formatting
print(f'Lower fence : {lower_fence:,.2f}')
print(f'Upper fence : {upper_fence:,.2f}')

# Detect and print outliers
for s in salaries:
    if s < lower_fence or s > upper_fence:
        print(f'🚨 Outlier detected: {s:,}')
````

## ⭐⭐ Task 5 — Z-Score Normalisation

**Dataset:** Heights (cm) of 10 people:
```python
heights = [150, 162, 170, 155, 180, 175, 168, 158, 172, 165]
```

**Your job:**
1. Compute the **mean** and **std** of heights
2. Calculate the **z-score** for each height
3. Print each height with its z-score
4. Flag anyone with **|z| > 1.5** as `'⚠️ Unusual'`

**Expected output:**
```
Height  Z-Score  Flag
 150     -1.84   ⚠️ Unusual
 162     -0.68
 170     +0.09
 ...
 180     +1.08
```

> 💡 **Hints:**
> - `z = (x - mean) / std`
> - Loop: `for h in heights:`
> - `flag = '⚠️ Unusual' if abs(z) > 1.5 else ''`
> - This is exactly what `StandardScaler` does in scikit-learn!

````python
# Task 5 — Your code here
import math

heights = [150, 162, 170, 155, 180, 175, 168, 158, 172, 165]

# Compute mean and std
mean = sum(heights) / len(heights)

# Calculate variance and standard deviation
variance = sum((h - mean) ** 2 for h in heights) / len(heights)
std = math.sqrt(variance)

# Print table header
print(f'  {"Height":>6}  {"Z-Score":>8}  Flag')
print('-' * 30)

# Calculate and print z-scores
for h in heights:
    z = (h - mean) / std
    flag = '⚠️ Unusual' if abs(z) > 1.5 else ''
    
    # Format z-score with sign (+ or -) and 2 decimal places
    z_formatted = f"{z:+.2f}"
    print(f'  {h:>6}  {z_formatted:>8}  {flag}')
````

## ⭐⭐ Task 6 — Explore Skewness

**Create these two datasets:**
```python
right_skewed = [2, 3, 3, 4, 4, 4, 5, 5, 10, 15, 20]
left_skewed  = [1, 5, 10, 15, 16, 16, 17, 17, 18, 18, 19]
```

**Your job:**
1. Compute **skewness** for both datasets using `stats.skew()`
2. Compute **mean** and **median** for both
3. Print results and verify the rule:
   - Right skew → mean **>** median
   - Left skew  → mean **<** median

**Expected output:**
```
--- Right Skewed ---
Skewness: +1.56   direction: positive →
Mean: 7.73   Median: 4.0
Mean > Median ✓  (right skew rule holds!)

--- Left Skewed ---
Skewness: -1.38   direction: negative ←
Mean: 13.0   Median: 16.0
Mean < Median ✓  (left skew rule holds!)
```

> 💡 **Hints:**
> - `stats.skew(data)` returns a number
> - Positive = right skew, Negative = left skew
> - Check the mean vs median relationship each time

````python
# Task 6 — Your code here
from scipy import stats

right_skewed = [2, 3, 3, 4, 4, 4, 5, 5, 10, 15, 20]
left_skewed  = [1, 5, 10, 15, 16, 16, 17, 17, 18, 18, 19]

# For right_skewed:
print("--- Right Skewed ---")
skew_right = stats.skew(right_skewed)
mean_right = sum(right_skewed) / len(right_skewed)
median_right = sorted(right_skewed)[len(right_skewed) // 2]

print(f"Skewness: {skew_right:+.2f}   direction: positive →")
print(f"Mean: {mean_right:.2f}   Median: {median_right:.1f}")

if mean_right > median_right:
    print("Mean > Median ✓  (right skew rule holds!)")
else:
    print("Mean < Median ✗  (rule doesn't hold)")
    
print()

# For left_skewed:
print("--- Left Skewed ---")
skew_left = stats.skew(left_skewed)
mean_left = sum(left_skewed) / len(left_skewed)
median_left = sorted(left_skewed)[len(left_skewed) // 2]

print(f"Skewness: {skew_left:+.2f}   direction: negative ←")
print(f"Mean: {mean_left:.2f}   Median: {median_left:.1f}")

if mean_left < median_left:
    print("Mean < Median ✓  (left skew rule holds!)")
else:
    print("Mean > Median ✗  (rule doesn't hold)")
````

## ⭐⭐ Task 7 — Correlation Analysis

**Dataset:** Hours of sleep and test scores for 10 students:
```python
hours_sleep = [5, 6, 6, 7, 7, 8, 8, 9, 9, 10]
test_score  = [55, 60, 62, 68, 71, 75, 78, 82, 85, 88]
```

**Your job:**
1. Calculate the **Pearson correlation** and **p-value**
2. Determine strength: Strong (>0.8) / Moderate (>0.5) / Weak
3. State if the result is statistically significant (p < 0.05)
4. **Bonus:** Plot a scatter plot with a trend line

**Expected output:**
```
Pearson r = 0.997
p-value   = 0.000001
Strength : STRONG positive correlation
Significant: YES ✅
```

> 💡 **Hints:**
> - `corr, pval = stats.pearsonr(x, y)`
> - Use `if/elif/else` for strength labels
> - Bonus: `plt.scatter(x, y)` then `np.polyfit(x, y, 1)` for trend line

````python
# Task 7 — Your code here
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

hours_sleep = [5, 6, 6, 7, 7, 8, 8, 9, 9, 10]
test_score  = [55, 60, 62, 68, 71, 75, 78, 82, 85, 88]

# 1. Pearson correlation
corr, pval = stats.pearsonr(hours_sleep, test_score)

print(f"Pearson r = {corr:.3f}")
print(f"p-value   = {pval:.6f}")

# 2. Strength classification
if abs(corr) > 0.8:
    strength = "STRONG"
elif abs(corr) > 0.5:
    strength = "MODERATE"
elif abs(corr) > 0.3:
    strength = "WEAK"
else:
    strength = "VERY WEAK"

direction = "positive" if corr > 0 else "negative"
print(f"Strength : {strength} {direction} correlation")

# 3. Significance check
significant = pval < 0.05
print(f"Significant: {'YES ✅' if significant else 'NO ❌'}")

# 4. (Bonus) Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(hours_sleep, test_score, color='blue', s=100, alpha=0.7)

# Add trend line
z = np.polyfit(hours_sleep, test_score, 1)
p = np.poly1d(z)
plt.plot(hours_sleep, p(hours_sleep), "r--", alpha=0.8, label=f'Trend line (r={corr:.3f})')

plt.xlabel('Hours of Sleep', fontsize=12)
plt.ylabel('Test Score', fontsize=12)
plt.title('Relationship Between Sleep and Test Scores', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
````

## ⭐⭐⭐ Task 9 — Hypothesis Testing

**Scenario:** A school tested a new teaching method.

```python
old_method = [65, 70, 68, 72, 74, 69, 71, 73, 66, 75]
new_method = [78, 82, 80, 85, 79, 83, 81, 84, 77, 86]
```

**Your job:**
1. State **H₀** and **H₁** in plain English (as print statements)
2. Run an **independent samples t-test**
3. Print t-statistic and p-value
4. Print the means of both groups
5. Write a clear **conclusion** based on the p-value

**Expected output:**
```
H₀: Both methods produce the same results
H₁: The new method is significantly different

Old method mean: 70.3
New method mean: 81.5
t-statistic    : -10.470
p-value        : 0.000000

✅ Reject H₀ — new method is significantly better!
```

> 💡 **Hints:**
> - `stats.ttest_ind(group1, group2)`
> - If p < 0.05 → reject H₀
> - Compare the means to see which group performed better
> - The t-stat being negative just means group1 < group2

````python
from scipy import stats
import numpy as np

old_method = [65, 70, 68, 72, 74, 69, 71, 73, 66, 75]
new_method = [78, 82, 80, 85, 79, 83, 81, 84, 77, 86]

# 1. State hypotheses
print("H₀: Both methods produce the same results (no significant difference)")
print("H₁: The new method results are significantly different from the old method")
print("-" * 30)

# 2. Run independent samples t-test
t_stat, p_val = stats.ttest_ind(old_method, new_method)

# 3. Print group means
old_mean = np.mean(old_method)
new_mean = np.mean(new_method)
print(f"Old method mean: {old_mean:.1f}")
print(f"New method mean: {new_mean:.1f}")

# 4. Print t-statistic and p-value
print(f"t-statistic    : {t_stat:.3f}")
print(f"p-value        : {p_val:.6f}")
print("-" * 30)

# 5. Conclusion
alpha = 0.05
if p_val < alpha:
    print("✅ Reject H₀ — The new method is significantly better!")
else:
    print("❌ Fail to reject H₀ — No significant difference found.")
````

````python
import numpy as np
from scipy import stats

def stats_report(data, name):
    # Convert to numpy array for easier math
    data = np.array(data)
    
    print('=' * 40)
    print(f'STATS REPORT: {name}')
    print('=' * 40)
    
    # 1. Values
    print(f"Values    : {data.tolist()}")

    # 2. Central Tendency
    mean = np.mean(data)
    median = np.median(data)
    mode_res = stats.mode(data, keepdims=True)
    mode_val = mode_res.mode[0]
    mode_count = mode_res.count[0]

    print("\n--- Central Tendency ---")
    print(f"Mean      : {mean:.1f}")
    print(f"Median    : {median:.1f}")
    print(f"Mode      : {mode_val} (appeared {mode_count}x)")

    # 3. Spread
    std_dev = np.std(data)
    variance = np.var(data)
    print("\n--- Spread ---")
    print(f"Std Dev   : {std_dev:.2f}")
    print(f"Variance  : {variance:.1f}")

    # 4. Quartiles & IQR
    q1, q2, q3 = np.percentile(data, [25, 50, 75])
    iqr = q3 - q1
    print("\n--- Quartiles ---")
    print(f"Q1={q1:<8} Q2={q2:<8} Q3={q3:<8} IQR={iqr}")

    # 5. Skewness
    skew = stats.skew(data)
    direction = "right skewed" if skew > 0 else "left skewed" if skew < 0 else "symmetric"
    print("\n--- Shape ---")
    print(f"Skewness  : {skew:+.2f} ({direction})")

    # 6. Outliers (IQR Method)
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    
    print("\n--- Outliers (IQR method) ---")
    if len(outliers) > 0:
        for val in outliers:
            print(f"🚨 Outlier: {val}")
    else:
        print("No outliers detected.")

    # 7. Z-scores
    print("\n--- Z-Scores ---")
    for val in data:
        z = (val - mean) / std_dev
        flag = "⚠️ EXTREME" if abs(z) > 2 else ""
        print(f"  {val:>3}   z={z:>+5.2f}  {flag}")

    print('=' * 40)

# Test it!
stats_report([12, 15, 14, 10, 100, 13, 14, 15, 16, 11], 'Sample Data')
````
